### Q3 빅데이터반의 10명의 학생이 있다. 중간고사 점수는 아래와 같다

[70, 60, 55, 75, 95, 80, 80, 85, 100, 90]  

### for문을 활용하여 빅데이터반의 평균을 구하세요

```python
# answer..
big_data = [70, 60, 55, 75, 95, 80, 80, 85, 100, 90]  

result = 0
for i in big_data:
    result += i
    
average1 = result/len(big_data)
print(average1)    
```

### Q4

```python
numbers = [1,2,3,4,5]
result = []
for i in numbers:
    if i%2 ==1:
        result.append(i*2)
```
