```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
x = np.arange(-5.0, 5.0, 0.1)
y = sigmoid(x)
```

```python
plt.plot(x,y,'g')
plt.plot([0,0],[1.0,0.0],':') #    
plt.title('sigmoid')
plt.show()
```

```python
# w  

y1 = sigmoid(0.5*x)
org_y = sigmoid(x)
y2 = sigmoid(2*x)

plt.plot(x,y1,'r', linestyle='--') # w 0.5 
plt.plot(x,org_y,'g')
plt.plot(x,y2,'b', linestyle='--') # w 2 
plt.plot([0,0],[1.0,0.0],':') #    
plt.title('sigmoid')
plt.show()
```

```python
# bias  

y11 = sigmoid(x+0.5)
y12 = sigmoid(x+1)
y13 = sigmoid(x+1.5)

plt.plot(x,y11,'r', linestyle='--') # bias 0.5 
plt.plot(x,y12,'g')
plt.plot(x,y13,'b', linestyle='--') # bias 1.5   
plt.plot([0,0],[1.0,0.0],':') #    
plt.title('sigmoid')
plt.show()
```

```python
# AND_gate 
```

```python
# y = w1 * x1 + w2 * x2 + b

def and_gate(x1, x2):
    w1 = 0.5
    w2 = 0.5
    b = -0.7
    result = w1 * x1 + w2 * x2 + b # .
    if result <=0:
        return 0
    else:
        return 1
```

```python
print(':', and_gate(0,0), and_gate(0,1), and_gate(1,0), and_gate(1,1))
```
