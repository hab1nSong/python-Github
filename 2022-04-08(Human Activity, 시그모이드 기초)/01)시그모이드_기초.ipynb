```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
x = np.arange(-5.0, 5.0, 0.1)
y = sigmoid(x)
```

```python
plt.plot(x,y,'g')
plt.plot([0,0],[1.0,0.0],':') # 가운데에 점선을 추가하는 식
plt.title('sigmoid')
plt.show()
```

```python
# w값에 따른 변화

y1 = sigmoid(0.5*x)
org_y = sigmoid(x)
y2 = sigmoid(2*x)

plt.plot(x,y1,'r', linestyle='--') # w값이 0.5일때 
plt.plot(x,org_y,'g')
plt.plot(x,y2,'b', linestyle='--') # w값이 2일때 
plt.plot([0,0],[1.0,0.0],':') # 가운데에 점선을 추가하는 식
plt.title('sigmoid')
plt.show()
```

```python
# bias값에 따른 변화

y11 = sigmoid(x+0.5)
y12 = sigmoid(x+1)
y13 = sigmoid(x+1.5)

plt.plot(x,y11,'r', linestyle='--') # bias에 0.5일때 빨간색
plt.plot(x,y12,'g')
plt.plot(x,y13,'b', linestyle='--') # bias에 1.5를 더했을때 파란색 
plt.plot([0,0],[1.0,0.0],':') # 가운데에 점선을 추가하는 식
plt.title('sigmoid')
plt.show()
```

## 단층 퍼셉트론 기초

```python
# AND_gate를 푼다
```

```python
# y = w1 * x1 + w2 * x2 + b

def and_gate(x1, x2):
    w1 = 0.5
    w2 = 0.5
    b = -0.7
    result = w1 * x1 + w2 * x2 + b # 다중선형회귀식.
    if result <=0:
        return 0
    else:
        return 1
```

```python
print('결과값:', and_gate(0,0), and_gate(0,1), and_gate(1,0), and_gate(1,1))
```

```python
# nand_gate를 푼다
```

```python
# y = w1 * x1 + w2 * x2 + b

def nand_gate(x1, x2):
    w1 = -0.5
    w2 = -0.5
    b = +0.7
    result = w1 * x1 + w2 * x2 + b # 다중선형회귀식.
    if result <=0:
        return 0
    else:
        return 1
```

```python
print('결과값:', nand_gate(0,0), nand_gate(0,1), nand_gate(1,0), nand_gate(1,1))
```

```python
print('결과값:', or_gate(0,0), or_gate(0,1), or_gate(1,0), or_gate(1,1))
```
