### X축 스케일 지정하기

#### matplotlib의 xscale(), yscale() 함수 활용 :: 축 스케일 지정

* 축의 기본 :: 'linear'스케일로 표시


* 'log','symlog','logit'으로 변경가능

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-10,10,100)
y = x**3

plt.plot(x,y)
plt.xscale('symlog')
plt.show()
```

### Y축 스케일 지정하기

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 5, 100)
y = np.exp(x)

plt.plot(x,y)
plt.yscale('linear')
#plt.yscale('log')
plt.show()
```
