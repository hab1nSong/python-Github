### 03)-1 기본사용

* x,y축에 대한 레이블 표시


* xlabel(),ylabel()함수를 활용 

```python
import matplotlib.pyplot as plt

plt.plot([1,2,3,4],[1,4,9,16])
plt.xlabel('Axis_X')
plt.ylabel('Axis_y')
plt.show()
```

### 03)-2 여백 지정하기

```python
plt.plot([1,2,3,4],[1,4,9,16])
plt.xlabel('Axis_X', labelpad=15) #labelpad=15pt 파라미터 값을 의미
plt.ylabel('Axis_y', labelpad=20) #labelpad=20pt 파라미터 값을 의미
plt.show()
```

### 03)-3 폰트 설정하기

```python
import matplotlib.pyplot as plt

plt.plot([1,2,3,4],[2,3,5,10])
plt.xlabel('Axis_X', labelpad=15, fontdict={'family':'serif',
                                           'color':'b',
                                           'weight':'bold',
                                           'size':14})
plt.ylabel('Axis_Y', labelpad=20, fontdict={'family':'fantasy',
                                           'color':'deeppink',
                                           'weight':'normal',
                                           'size':'xx-large'})
plt.show()
```

```python
# 변수에 대입해서 사용하는 방법

import matplotlib.pyplot as plt

font1 = {'family':'serif',
        'color':'b',
        'weight':'bold',
        'size':14}

font2 = {'family':'fantasy',
        'color':'deeppink',
        'weight':'normal',
        'size':'xx-large'}

plt.plot([1,2,3,4],[2,3,5,10])
plt.xlabel('Axis_X', labelpad=15, fontdict=font1)
plt.ylabel('Axis_Y', labelpad=20, fontdict=font2)
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.plot([1,2,3,4],[2,3,5,10])
plt.xlabel('Axis_X', loc='right')
plt.ylabel('Axis_y', loc='top')
plt.show()
```
