## 04. Matplotlib 범례 표시하기

### 기본 사용 

* 범례(Legend)는 그래프에 데이터의 종류를 표시하기 위함

* matplotlib.pyplot모듈의 legend()함수를 통해 범례 표현

```python
import matplotlib.pyplot as plt

plt.plot([1,2,3,4], [2,3,5,10], label='Price( $ )')
plt.xlabel('Axis_X')
plt.ylabel('Axis_y')
plt.legend()

plt.show()
```

### 위치 지정하기

```python
import matplotlib.pyplot as plt

plt.plot([1,2,3,4],[2,3,5,10], label='Sales')
plt.xlabel('Axis_X')
plt.ylabel('Axis_y')
plt.legend(loc=(0.4, 0.5))

plt.show()
```

```python
#2번째 방법

import matplotlib.pyplot as plt

plt.plot([1,2,3,4],[2,3,5,10], label='Sales')
plt.xlabel('Axis_X')
plt.ylabel('Axis_y')
plt.legend(loc='lower right')

plt.show()
```

### 열 개수 지정하기

```python
plt.plot([1,2,3,4],[2,3,5,10],label='Price')
plt.plot([1,2,3,4],[3,5,9,7], label='Demand')
plt.xlabel('Axis_X')
plt.ylabel('Axis_y')
#plt.legend(loc='best')  # ncol = 1
plt.legend(loc='best', ncol=2)

plt.show()
```

### 폰트 크기 지정하기

```python
import matplotlib.pyplot as plt

plt.plot([1, 2, 3, 4], [2, 3, 5, 10], label='Price ($)')
plt.plot([1, 2, 3, 4], [3, 5, 9, 7], label='Demand (#)')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
# plt.legend(loc='best')
plt.legend(loc='best', ncol=2, fontsize=14)

plt.show()
```

### 범례 테두리 꾸미기

```python
import matplotlib.pyplot as plt

plt.plot([1, 2, 3, 4], [2, 3, 5, 10], label='Price ($)')
plt.plot([1, 2, 3, 4], [3, 5, 9, 7], label='Demand (#)')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
# plt.legend(loc='best')
plt.legend(loc='best', ncol=2, fontsize=14, frameon=True, shadow=True)

### 폰트 크기 지정하기plt.show()
```
