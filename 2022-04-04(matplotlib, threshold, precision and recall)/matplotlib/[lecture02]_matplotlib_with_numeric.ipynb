## 01) play with numeric in Matplotlib

## If you need to draw Line or Marker :: plot()

```python
import matplotlib.pyplot as plt

plt.plot([2,3,5,10])
plt.show()
```

```python
# 만약 하나의 list 혹은 tuple 혹은 Numpy ndarray의 형태로도 입력이 가능
```

## X, y를 지정하여 입력

```python
plt.plot([1,2,3,4], [2,3,5,10]) # [1,2,3,4] :: x_values
                                # [2,3,5,10] :: y_values
plt.show()
```

```python
dict_data = {'ftr':[1,2,3,4,5],
            'data_y':[1,3,4,11,7]}

plt.plot('ftr','data_y', data=dict_data)
plt.show()
```

## 02) Matplotlib 축 레이블 설정하기

### 02)-1 기본사용 

```python
plt.plot([1,2,3,4],[1,4,9,16])
plt.xlabel('X-Label')
plt.ylabel('Y-Label')
plt.show()
```

### 02)-2 여백 지정하기

```python
plt.plot([1,2,3,4,5],[2,4,8,10,12])
plt.xlabel('X-Axis',labelpad=15) # 여백 지정하기
plt.ylabel('Y-Axis',labelpad=20) # 여백 지정하기
plt.show()
```

### 02)-3 폰트 설정하기

```python
plt.plot([1,2,3,4,5],[2,4,8,10,12])
plt.xlabel('X-Axis',labelpad=15, fontdict={'family':'serif',
                                          'color':'b',
                                          'weight':'bold',
                                          'size':14}) # font 지정하기

plt.ylabel('Y-Axis',labelpad=20, fontdict={'family':'serif',
                                          'color':'deeppink',
                                          'weight':'normal',
                                          'size':'xx-large'}) # font 지정하기
plt.show()
```

### 02)-4 위치 지정하기

```python
plt.plot([1,2,3,4,5],[2,4,8,10,12])
plt.xlabel('X-Axis',loc='right') # xlabel_loc 파라미터는 ({'left','center','right'})
plt.ylabel('Y-Axis',loc='bottom') # ylabel_loc 파라미터는 ({'bottom','center','top'})
plt.show()
```
