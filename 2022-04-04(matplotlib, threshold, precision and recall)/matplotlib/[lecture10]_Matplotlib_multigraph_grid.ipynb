### 10)-1여러 곡선 그리기

### 기본사용

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0, 2, 0.2)

plt.plot(x, x, 'r--', x, x**2, 'bo', x, x**3, 'g-.')
plt.show()
```

* plot()함수에 x값, y값, 스타일을 차례대로 세 번 입력


* y=x, y=x^2, y=x^3 

> 'r-' :: 빨간색(Red)의 대쉬(Dash) 스타일 선
>
> 'bo' :: 파란색(Blue)의 원형(Circle) 마커
>
> 'g-' :: 녹색(Green)의 대쉬-닷(Dash-dot) 스타일 선

### 10)-2 Matplotlib_grid

### 기본사용

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0, 2, 0.2)

plt.plot(x, x, 'bo')
plt.plot(x, x**2, color='#e35f62', marker='*', linewidth=2)
plt.plot(x, x**3, color='springgreen', marker='^', markersize=9)
plt.grid(True)

plt.show()
```
