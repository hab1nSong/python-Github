1.손실함수(Loss), 옵티마이저(Optimizer), 에포크(Epoch)의 개념

$$
\text{MSE}(\hat{x}_{1:N}, x_{1:N})=\frac{1}{N\times{n}}\sum_{i=1}^{N}{||x_i-\hat{x}_i||_2^2}
$$

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
# 1) MSE(Mean Squared Error, MSE)
model = Sequential()

model.compile(optimizer = 'adam', loss='mse', metrics=['mse'])

# == model.compile(optimzier = 'adam', loss=tf.keras.losses.MeanSquaredError(), metrics=['mse'])
```

```python
# 2) Binary cross-entropy :: 이항 교차 엔트로피 
# binary class에서의 entropy 적용

model.compile(loss = 'binary_crossentropy', optimizer = 'adam', metrics=['acc'])
```

```python
# 3) Categorical Cross-entropy :: 범주형 교차 엔트로피 (Multi class)

model.compile(loss='categorical_crossentropy', optimizer= 'adam', metrics=['acc'])
# == model.compile(optimzier = 'adam', loss=tf.keras.losses.CategoricalCrossentropy(), metrics=['acc'])
```

```python
# 3)-1 label에 대한 oh_encoding을 생략하고 진행한다면?

model.compile(loss='spars_categorical_crossentropy', optimizer= 'adam', metrics=['acc'])
# == model.compile(optimzier = 'adam', loss=tf.keras.losses.SparseCategoricalCrossentropy(), metrics=['acc'])
```
