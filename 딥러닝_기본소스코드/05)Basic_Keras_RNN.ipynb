1.전처리(Preprocessing)

```python
from typing import Sequence
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.layers import Embedding

tokenizer = Tokenizer()
train_text = 'The earth is a wonderful place live'

# 단어 집합 생성
tokenizer.fit_on_texts([train_text])

# int encoding
sub_text = "The earth is a awesome place live"
sequence = tokenizer.texts_to_sequences([sub_text])[0]

print('단어 집합:', tokenizer.word_index)
print('정수 인코딩:', sequence)

# vocabulary는 단어 집합에 존재하지 않으므로 출력되지 않음
```

```python
# pad_sequence 
pad_sequences([[1,2,3],[3,4,5,6],[7,8]])

# if you want to change maxlen...
pad_sequences([[1,2,3],[3,4,5,6],[7,8]], maxlen=3, padding='pre')
# first arg = data
# max_len = length of normalization
# padding = 'pre' or 'post'
```

4. 컴파일(Compile)&훈련(Training)

```python
# example of RNN

from tensorflow.keras.layers import SimpleRNN, Embedding, Dense
from tensorflow.keras.models import Sequential

vocab_size = 10000
embedding_dim = 32
hidden_units = 32

# design the model
model = Sequential()
model.add(Embedding(vocab_size, embedding_dim))
model.add(SimpleRNN(hidden_units))
model.add(Dense(1,activation='sigmoid'))

## Compile
model.compile(optimizer = 'rmsprop', loss = 'binary_crossentropy', metrics=['acc'])

## if you already split the data 
# model.fit(X_train, y_train, epochs= 10, batch_size=32, verbose = 0, validation_data(X_val,y_val))

## Evaluation and Prediction
# model.evaluate(X_test, y_test, batch_size = 32)

## Prediction
# model.predict(X_input, batch_size=32)

## Save and Load
# model.save('model_name.h5') 

# from tensorflow.keras.models import load_model
# model = load_model('model_name.h5')
```
