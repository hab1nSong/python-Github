```python
import numpy as np
import pandas as pd
```

```python
data = {'Name':['Chulmin','Eunkyoung','Jiwoong','Soobeom'],
        'Year':[2011,2016,2015,2015],
        'Gender':['Male','Female','Male','Male'],
        }

data_df = pd.DataFrame(data,index=['one','Two','Three','Four'])
data_df
```

```python
# 특정한 데이터 추출
print(data_df.iloc[0,0]) # 위치기반 인덱싱(positional index)
print(data_df.iloc[3,0]) # 위치기반 인덱싱(positional index)
print(data_df.iloc[[0,2],[0,1]]) # 위치기반 인덱싱(positional index)
print(data_df.iloc[:,:]) # 위치기반 인덱싱(positional index)
```

```python
# loc 명칭기반 인덱싱
print(data_df.loc[:,'Gender']) # 명칭기반 인덱싱(label based indexing)
print('\n',data_df.loc[data_df['Gender']=='Female']) # 명칭기반 인덱싱(label based indexing)
print('\n',data_df.loc[data_df['Year']<=2015]) # 명칭기반 인덱싱(label based indexing)
```

```python
# Q4. 나이가 60세 이상이고 선실등급이 1등급이며 성별이 여성인 승객을 추출해보세요
```

```python
q4_answer = titan_df[(titan_df.Age>=60)&(titan_df.Pclass==1)&(titan_df.Sex=='female')]
```

```python
# Q4. 나이가 60세 이상이고 선실등급이 1등급이며 성별이 여성인 승객을 추출해보세요-2
cond1 = titan_df.Age>=60
cond2 = titan_df.Pclass==1
cond3 = titan_df.Sex=='female'

titan_df[cond1&cond2&cond3][['Name']]
```

```python
# Q5. 위의 조건 전부를 부정하는 형태로 코드를 짜보세요.
titan_df[~((titan_df.Age>=60)&(titan_df.Pclass==1)&(titan_df.Sex=='female'))]
```

## 정렬, Aggregation, GroupBy

```python
titan_df.head()
titan_df['Name'].sort_values() # series의 경우
titan_df.sort_values(by=['Name']).head(3) # df의 경우
```

```python
# 여러개의 기준을 설정해서 정렬 :: 1차 오름차순- Pclass, 2차 오름차순-Name
titan_df_sorted = titan_df.sort_values(by=['Pclass','Name'])
titan_df_sorted

# 내림차순은? 
titan_df.sort_values(by=["Name"],ascending=False).head(2)
```

```python
## 참고 넘파이에서의 sort
array1 = np.array([3,1,5,9])
np.sort(array1)

# 기초 sort함수
org_array = array1.copy()
org_array.sort()
print(org_array) # 원본이 바뀐다 
```

### groupby()적용

```python
titan_groupby = titan_df.groupby(by='Pclass')[['Age','Cabin','Embarked']].count()
titan_groupby
```

```python
# age와 fare의 min과 max를 pclass의 unique한 value별로 보고싶다
titan_df.groupby(by=['Pclass'])[['Age','Fare']].agg(['min','mean','max'])
```
