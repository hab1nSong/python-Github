```python
import pandas as pd
stock_df = pd.read_csv('./stock_price.txt')
```

```python
stock_df.info()
```

```python
# q1. Yr_Mo_Dy을 판다스에서 인식할 수 있는 datetime64로 변경해보세요.

stock_df.Yr_Mo_Dy = pd.to_datetime(stock_df.Yr_Mo_Dy)
stock_df
```

```python
# q2. Yr_Mo_Dy에 존재하는 년도의 유일값을 모두 출력해보세요

stock_df.Yr_Mo_Dy.dt.year.unique()
```

```python
# q3. 2061이상의 데이터의 경우에는 잘못된 데이터이다. 그래서 해당 데이터의 경우
# 100을 빼서 새롭게 날짜를 assign하여서 Yr_Mo_Dy 컬럼에 정의해보세요..

def fix_year(x):
    import datetime
    
    year = x.year - 100 if x.year>= 2061 else x.year
    return pd.to_datetime(datetime.date(year, x.month, x.day))

stock_df['Yr_Mo_Dy'] = stock_df.Yr_Mo_Dy.apply(fix_year)
```

```python
# q4. 년도별 각 컬럼의 평균값을 구해보세요~
stock_df.groupby(stock_df.Yr_Mo_Dy.dt.year).mean().head(3)
```

```python
# q5. weekday컬럼을 만들고 요일별로 매핑해보겠습니다.(0~6:월요일~일요일)
stock_df['week_day'] = stock_df.Yr_Mo_Dy.dt.weekday
stock_df[['Yr_Mo_Dy','week_day']]
```

```python
# q6. week_day 컬럼을 기준으로 평일이면 0 주말이면 1인 새로운 컬럼 week_cat을 생성해보세요.

stock_df['week_cat'] = stock_df['week_day'].apply(lambda x: 1 if x in [5,6] else 0)
stock_df.head()
```

```python
# q7. 년도와 일자에 관련없이 모든 컬럼의 값들을 각 월(month)의 평균을 구한다면 어떻게 할까요?
stock_df.groupby(stock_df.Yr_Mo_Dy.dt.month).mean()
```

```python
# q8. 결측치를 확인하고 만약에 존재한다면, 컬럼기준의 직전(즉, t의 시점이라면, t-1의 시점의)의 값으로
# 대체하고 첫 번째 행에 결측값이 있다면 뒤애ㅔ 있는 값으로 대체해보세요..

stock_df.isna().sum() # 결측값을 확인~
# 결측치 대체(imputation)
stock_df = stock_df.fillna(method='ffill').fillna(method='bfill')
stock_df.isna().sum()
```

```python
# q9. RPT와 VAL의 컬럼을 일주일 간격으로 각각 Moving average 해보세요.
stock_df[['RPT','VAL']].rolling(7).mean().head(15)
```
