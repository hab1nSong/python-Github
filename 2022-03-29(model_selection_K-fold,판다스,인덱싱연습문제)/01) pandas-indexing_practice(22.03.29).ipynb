```python
# q1. 데이터를 로드하세요 (chipo.tsv)
import pandas as pd
import numpy as np
chip_df = pd.read_csv('D:/pythonTest/chipotle.tsv', delimiter='\t')
chip_df
```

```python
# q2. 로딩된 데이터의 info를 확인하기.
chip_df.info()
```

```python
# q3. quantity컬럼 값이 3인 데이터를 추출하고 index를 0부터 정렬하고 첫 3행 출력해보세요.
chip_df[chip_df.quantity==3].reset_index(drop=True).head(3)
```

```python
# q4. quantity와 item_price 두개의 컬럼으로 구성된 새로운 df를 생성하세요.
chip_ndf = chip_df[['quantity','item_price']]
chip_ndf
```

```python
# q5. item_price에서 $형식에 숫자가 붙여진 (예: $2.39, $3.39)데이터에서 $를 제외시키고
# str이었던 데이터 타입을 float으로 변경하라
# 변경된 데이터를 new_price라고 하는 컬럼을 생성하여
# 여러분의 df에 입력하세요.
chip_item = chip_df['item_price'].str.replace('$','')
chip_item_price = chip_item.astype('float64')
chip_item_price.dtype
chip_df['new_price'] = chip_item_price
chip_df

# 다른 방법
cond1 = chip_df.item_price.str[1:]
result=cond1.astype('float64')
chip_df['new_price'] = result
chip_df.head()

# 위에 코드를 한줄로 바꾸면?
chip_df['new_price'] = chip_df.item_price.str[1:].astype('float64')
chip_df
```

```python
# q6. 위에서 생성된 new_price의 컬럼에서 5이하의 값을 갖는 df를 추출하고, 전체의 개수도 구해보세요.

chip_new_price = chip_df.loc[chip_df['new_price']<=5]
print(chip_new_price)
chip_new_price.shape

# 강사님의 풀이
len(chip_df[chip_df.new_price<=5])
chip_df[chip_df.new_price<=5].shape[0]
```

```python
# q8. new_price값이 9이하이고 item_name값이 Chicken Salad Bowl인 df를 추출해보세요.
chip_df[(chip_df.new_price<=9)&(chip_df.item_name=='Chicken Salad Bowl')].head()
```

```python
# q16. df의 데이터 중에서 choice_description의 값의 결측값을 발견했다면
## 위의 데이터를 NoData로 변경해보세요...(loc를 활용하는 방법도 있어요)

chip_df['choice_description'].fillna('NoData') #  fillna를 활용하는 방법 :: inplace = True등을 기억
chip_df.loc[chip_df.choice_description.isna(), 'choice_description'] = 'NoData' # 2번째 방법
chip_df[chip_df.choice_description=='NoData'].shape[0]
```
