### bool 자료형

bool 자료형이란 참(True)과 거짓(False)을 나타내는 자료형이다. 따라서, bool 자료형은 2가지 값만을 가질 수 있다. 

* True - 참

* False - 거짓

```python
# a= True, b= False 지정해봐욤

a= True
b= False

print(type(a))
print(type(b))
```

### 자료형의 참과 거짓

```python
for i in [1,2,3]:
    print(i)
```

```python
a= [1,2,3]
while a :
     print(a.pop())
```

```python
if []:
    print('참')
else:
    print('거짓')
```

```python
# 불 연산을 보고 싶다면?
bool('python')
bool('')
bool([1,2,3]) # :: true
bool(0) # :: false
bool([]) #:: false
```
